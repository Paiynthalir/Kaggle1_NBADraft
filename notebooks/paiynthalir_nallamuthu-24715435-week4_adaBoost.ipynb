```python
# 1.2. Import the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
```

```python
# 1.3. Load the training, validation and test datasets into a dataframes 
# Load the Parquet files into DataFrames
X_train = pd.read_parquet('../data/processed/X_train.parquet')
X_val = pd.read_parquet('../data/processed/X_val.parquet')
y_train = pd.read_parquet('../data/processed/y_train.parquet')
y_val = pd.read_parquet('../data/processed/y_val.parquet')
X_test = pd.read_parquet('../data/processed/X_test_scaled.parquet')
```

```python
# taking out the player id out of the test data
player_id_test = X_test.pop('player_id')
```

```python
y_val = y_val.iloc[:, 0]  # Convert the single column DataFrame to a Series
y_val.shape
```

```python
y_train = y_train.iloc[:, 0]  # Convert the single column DataFrame to a Series
y_train.shape
```

3. Ada Boost - Modelling

```python
#3.1 Import adaBoost and instantiate the model
# Initialize the ada boost model
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# Initialize the AdaBoost Classifier models
ada1 = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1), 
    n_estimators=200,          
    learning_rate=0.05,        
    random_state=0,
    algorithm="SAMME"
)

ada2 = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1), 
    n_estimators=200,          
    learning_rate=0.05,        
    random_state=0,
    algorithm="SAMME"
)
```

```python
# 3.2 Train the model and save the predicted values
# Fit the AdaBoost model
ada2.fit(X_train, y_train)

# Get feature importances
feature_importances = ada2.feature_importances_

# Create a DataFrame to hold the features and their importances
feat_imp_df = pd.DataFrame({
    'feature': X_train.columns,
    'importance': feature_importances
})

# Select features above the importance threshold
important_features = feat_imp_df[feat_imp_df['importance'] >= 0.02]['feature']
```

```python
# Transform the training and validation sets to include only the selected features
X_train_impFeatures = X_train[important_features]
X_val_impFeatures = X_val[important_features]
X_test_impFeatures = X_test[important_features]

# fit the model on the selected features
ada2.fit(X_train_impFeatures, y_train)
```

```python
#Fit the base model without feature importance 
ada1.fit(X_train, y_train)
```

```python
# Select the probability for the positive class (class 1) - adaBoost with out feature importance
y_train_probs_ada1 = ada1.predict_proba(X_train)[:, 1]
y_val_probs_ada1 = ada1.predict_proba(X_val)[:, 1]

# Select the probability for the positive class (class 1) - adaBoost with feature importance
y_train_probs_ada2 = ada2.predict_proba(X_train_impFeatures)[:, 1]
y_val_probs_ada2 = ada2.predict_proba(X_val_impFeatures)[:, 1]
```

```python
# 3.3 To plot ROC curve and the AUC metric - for comparison of model performances
#To plot ROC curve and the AUC metric
from sklearn import metrics

# Set up plotting area
plt.figure(0).clf()

# Gradient Boost- Train
fpr, tpr, _ = metrics.roc_curve(y_train, y_train_probs_ada1)
auc = round(metrics.roc_auc_score(y_train, y_train_probs_ada1), 4)
plt.plot(fpr, tpr, label="Ada Boost Train, AUC="+str(auc))

# Gradient Boost - Validation
fpr, tpr, _ = metrics.roc_curve(y_val, y_val_probs_ada1)
auc = round(metrics.roc_auc_score(y_val, y_val_probs_ada1), 4)
plt.plot(fpr, tpr, label="Ada Boost Val, AUC="+str(auc))

# Gradient Boost with important featues- Train
fpr, tpr, _ = metrics.roc_curve(y_train, y_train_probs_ada2)
auc = round(metrics.roc_auc_score(y_train, y_train_probs_ada2), 4)
plt.plot(fpr, tpr, label="adaBoost with impFeatures Train, AUC="+str(auc))

# Gradient Boost with important featues - Validation
fpr, tpr, _ = metrics.roc_curve(y_val, y_val_probs_ada2)
auc = round(metrics.roc_auc_score(y_val, y_val_probs_ada2), 4)
plt.plot(fpr, tpr, label="adaBoost with impFeatures Val, AUC="+str(auc))

# Add legend
plt.legend()

# Show the plot
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```

```python
# 3.4 Save the probability of the test data
y_test_probs_ada1 = ada1.predict_proba(X_test)[:, 1]
```

```python
# 3.5 add the 'player_id'column  into the prediction probability
# Convert y_test_prob_rf3 (which is a NumPy array) to a DataFrame
y_test_probs_df = pd.DataFrame(y_test_probs_ada1, columns=['drafted'])

# Concatenate player_id_test and y_test_prob_df along the columns
Results = pd.concat([player_id_test, y_test_probs_df], axis=1)
```
